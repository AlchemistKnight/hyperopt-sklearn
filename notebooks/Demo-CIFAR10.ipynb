# Hyperopt-Sklearn on CIFAR-10

The CIFAR-10 object recognition data set is a follow-up to the well-known MNIST data set of hand-written digits. CIFAR-10 is a data set of 60 000 small 32x32 color images of every-day objects. There are 10 kinds of object in CIFAR-10, and the data set is typically used for 10-way classification.




```python
# -- RETRIEVE DATA VIA SKDATA (github.com/jaberg/skdata)
from skdata.cifar10 import (view, dataset)
dataset.CIFAR10().fetch(download_if_missing=True, verbose=False)
dataview = view.OfficialVectorClassificationTask(x_dtype='float32')
```

```python
import hpsklearn
import hyperopt.tpe
estimator = hpsklearn.HyperoptEstimator(
    #preprocessing=simple_small_image_preprocessing('pp'),
    #classifier=hpc.any_classifier('classif'),
    max_evals=100,
    verbose=1,
    algo=hyperopt.tpe.suggest,
    trial_timeout=60.0 * 10, # -- seconds
    )
```

```python
# Demo version of estimator.fit()
import hpsklearn.demo_support
reload (hpsklearn.demo_support)

fit_iterator = estimator.fit_iter(dataview.train.x, dataview.train.y)
fit_iterator.next()
plot_helper = hpsklearn.demo_support.PlotHelper(estimator)
while len(estimator.trials.trials) < estimator.max_evals:
    fit_iterator.send(1) # -- try one more model
    plot_helper.post_iter()
plot_helper.post_loop()

# -- Model selection was done on a subset of the training data.
# -- Now that we've picked a model, train on all training data.
estimator.retrain_best_model_on_full_data(dataview.train.x, dataview.train.y)
```
