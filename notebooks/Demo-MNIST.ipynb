# Hyperopt-Sklearn on MNIST

The MNIST data set of hand-written digits has 70 000 small 28x28 gray-scale images of the digits 0-9.
The data set is typically used for 10-way classification.

Professor Yann LeCun maintains a list of "high-scores" for this dataset. The best performers are convolutional neural networks trained with an artificially augmented training set. Such variations on such methods score between 99% and 100% accuracy. The simpler classification approach respresented by hyperopt-sklearn scores a respectable 98.5%. The search below takes about one hour of total wall time.

```python
# -- RETRIEVE DATA VIA SKDATA (github.com/jaberg/skdata)
from skdata.mnist import view
dataview = view.OfficialVectorClassification(x_dtype='float32')
```

```python
import hpsklearn
import hyperopt.tpe
estimator = hpsklearn.HyperoptEstimator(
    #preprocessing=simple_small_image_preprocessing('pp'),
    #classifier=hpc.any_classifier('classif'),
    max_evals=100,
    verbose=1,
    algo=hyperopt.tpe.suggest,
    trial_timeout=60.0 * 10, # -- seconds
    )
```

```python
# Demo version of estimator.fit()
import hpsklearn.demo_support

fit_iterator = estimator.fit_iter(dataview.all_vectors[dataview.sel_idxs],
                                  dataview.all_labels[dataview.sel_idxs])
fit_iterator.next()
plot_helper = hpsklearn.demo_support.PlotHelper(estimator,
                                                mintodate_ylim=(-0.001, 0.025),
                                                )
while len(estimator.trials.trials) < estimator.max_evals:
    fit_iterator.send(1) # -- try one more model
    plot_helper.post_iter()
plot_helper.post_loop()

# -- Model selection was done on a subset of the training data.
# -- Now that we've picked a model, train on all training data.
estimator.retrain_best_model_on_full_data(dataview.all_vectors[dataview.sel_idxs],
                                          dataview.all_labels[dataview.sel_idxs])
```
